#2.날짜별 총 승객수 추이(lineplot 사용)


```python
plt.figure(figsize=(14, 7))
daily_ridership = subway_df.groupby('사용일자')['승차총승객수'].sum().reset_index()
sns.lineplot(x='사용일자', y='승차총승객수', data=daily_ridership)
plt.title('날짜별 총 승객수 추이')
plt.xlabel('사용일자')
plt.ylabel('승차총승객수')
plt.xticks(rotation=45)
plt.show()
```

#3. 노선별 승차 총승객수와 하차총승객수 비교(barplot사용)



```python
plt.figure(figsize=(14, 7))
line_ridership = subway_df.groupby('노선명')[['승차총승객수', '하차총승객수']].sum().reset_index()
line_ridership = line_ridership.melt(id_vars='노선명', value_vars=['승차총승객수', '하차총승객수'], var_name='종류', value_name='승객수')
sns.barplot(x='노선명', y='승객수', hue='종류', data=line_ridership)
plt.title('노선별 승차총승객수와 하차총승객수 비교')
plt.xticks(rotation=45)
plt.show()
```

#12. 승하차총승객수 상위 10개 역의 파이차트

```python
top_10_stations = subway_df.groupby('역명')[['승차총승객수', '하차총승객수']].sum().nlargest(10, '승차총승객수')
top_10_stations['총승객수'] = top_10_stations['승차총승객수'] + top_10_stations['하차총승객수']

plt.figure(figsize=(12, 8))
top_10_stations['총승객수'].plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('승하차총승객수 상위 10개 역의 파이차트')
plt.ylabel('')
plt.show()
```

```python
specific_station = '서울역'  # 예시 역
station_data = subway_df[subway_df['역명'] == specific_station]

plt.figure(figsize=(14, 7))
station_data['총승객수'] = station_data['승차총승객수'] + station_data['하차총승객수']
sns.lineplot(x='사용일자', y='총승객수', data=station_data)
plt.title(f'{specific_station}의 일자별 승하차총승객수 합계 추이')
plt.xlabel('사용일자')
plt.ylabel('총승객수')
plt.xticks(rotation=45)
plt.show()
```
